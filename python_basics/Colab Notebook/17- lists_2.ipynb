```python
#Cópia
```

```python
n = [11, 22, 33, 44, 55]
```

```python
n2 = n
```

```python
n2.append(1)
```

```python
n2
```

```python
n
```

```python
n3 = n[:]
```

```python
n3
```

```python
n3.append(8)
```

```python
n3
```

```python
n
```

```python
''' lista = lista2 -> referência, tudo que substitui em uma, 
substitui na outra
    lista = lista2[:] -> cópia dos elementos de lista2 para
lista 1, dessa forma as listas são independentes'''
```

```python
n4 = n3.copy()
```

```python
n4.append(2)
```
