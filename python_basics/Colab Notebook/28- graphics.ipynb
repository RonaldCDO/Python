```python
x = np.linspace(-1,1,100)
```

```python
y = 2*x
y2 = x**2
y3 = x**3
y4 = np.sin(x)
```

```python
fig, janela = plt.subplots(2,2)

janela[0,0].plot(x,y,'.')
janela[0,1].plot(x,y2,'.g')
janela[1,0].plot(x,y3,'.y')
janela[1,1].plot(x,y4,'.r')

fig.legend(['Reta', 'Parábola', 'Cúbica','Sin'])
plt.show()
```
