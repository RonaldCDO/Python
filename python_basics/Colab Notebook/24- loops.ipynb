```python
#Pass, break & continue
```

```python
for letra in 'analista':
  if letra == 'a':
    pass
  print(letra)
```

```python
for letra in 'analista':
  if letra == 'a':
    continue
  print(letra)
```

```python
for letra in 'analista':
  if letra == 'a':
    break
  print(letra)
```
