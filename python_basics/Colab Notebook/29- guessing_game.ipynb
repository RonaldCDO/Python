```python
import random
```

```python
def level ():
  boolAux = True
  while boolAux:
    try:  
      print("Select the level of the Guessing Game 1 for easy, 2 for medium and 3 for hard: ")
      level = int(input())
    except ValueError:
      print ("Invalid Input")
      continue

    if level == 1 or level == 2 or level == 3:
      boolAux = False
    else:
      print("Invalid input! Type the level of your choice")

    if level == 1:
      chances = 10
    elif level == 2:
      chances = 6
    elif level == 3:
      chances = 3
  return chances
```

```python
def guessGame():
  targetNumber = random.randint(0,50)  

  chances = level()

  for tries in range(chances,-1,-1):
    
    if(tries == 0):
      print("You Lost! The correct Number is ", targetNumber)
      break
    
    if(tries == 1):
      print("Last Chance")
    else:
      print("You have ", tries, "chances to guess a number between 0 and 50!")
    
    try:
      print("Guess the number: ")
      guessNumber = int(input())
    except ValueError:
      print("Invalid input! The number must be an int from range 0 to 50")
      continue

    if guessNumber == targetNumber:
      print("Correct Answer! The target number is: ", targetNumber)
      break
    elif guessNumber > targetNumber:
      print("The Guessed number is bigger than the target Number!")
    else:
      print("The Guessed number is lesser than the target Number!")
```

```python
guessGame()
```
